```python
import numpy as np
from scipy.optimize import linprog

# Example data (same as section 2.5)
p = np.array([2.0, 3.0])       # prices
S = np.array([[1.0, 0.0],
              [0.0, 2.0],
              [1.0, 1.0]])    # m x d
X = np.array([1.0, 2.0, 1.0])

# linprog expects form: minimize c^T x subject to A_ub x <= b_ub
# Our constraints: S theta >= X  -> -S theta <= -X
c = p
A_ub = -S
b_ub = -X

res = linprog(c, A_ub=A_ub, b_ub=b_ub, method='highs')
print("Status:", res.message)
print("Optimal cost:", res.fun)
print("theta:", res.x)
```
